```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/data cleaing/employee_data (1).csv")
```

```python
df.info()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df
df['Employee Name']= df['Employee Name'].fillna("unknown")
df['Department']= df['Department'].fillna("unknown")
df['Join Date']= df['Join Date'].fillna(df['Join Date'].mode()[0])
print(df.isnull().sum())
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
df = df.drop_duplicates()
print(df.head())
```

```python
print(df['Employee Name'].head())
print(df['Department'].head())
```

```python
df['Employee Name'] = df['Employee Name'].str.strip().str.title()

df['Department'] = df['Department'].str.strip().str.title()
print(df[['Employee Name', 'Department']].head())
```

```python
print("Before:")
print(df['Department'].unique())
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
)
print(df.columns)
```

```python
df['join_date'] = pd.to_datetime(df['join_date'], format='mixed')
print(df['join_date'].unique())
```
