```python
import seaborn as sns
df =sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.isnull().sum()
#
```

```python
df.shape
```

```python
df["age"]
```

```python
df.columns
```

```python
df["Age_mean"] = df["age"].fillna(df["age"].mean())
print(df["Age_mean"])
```

```python
df[["Age_mean", 'age']]
```

```python
df["Age_Median"] = df["age"].fillna(df["age"].median())
print(df["Age_median"])
```

```python
df[["Age_Median", 'age']]
```

```python
df["Mode"] = df["age"].fillna(df["age"].mode()[0])
print(df["Mode"])
```
