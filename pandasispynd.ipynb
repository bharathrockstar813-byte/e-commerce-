```python
#series
arr = np.array([10,20,30,40,50])
marks = pd.Series(arr)
print(marks)
```

```python
marks = pd.Series(
    [10,20,30],
    index = ["math","phy","chem"]
)
print(marks)
print(marks["math"])
print(marks.shape)
print(marks.size)
print(marks.dtype)
print(marks.index)
```
