```python
import numpy as np
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([10,20,30])
print(arr1+arr2)
print(arr2-arr1)
print(arr1*arr2)
print(arr2/arr1)
print(arr1%arr2)
print(arr1**arr2)
```

```python
l = np.array([10,20,30,40,50,60])
#add elements = sum
print(np.sum(l))
#average
print(np.mean(l))
#max
print(np.max(l))
#min
print(np.min(l))
#stand devetion
print(np.std(l))
#variance
print(np.var(l))
#size of element in an array
print(np.size(l))
#arg max
print(np.argmax(l))
#arg min
print(np.argmin(l))
#applaying square root
print(np.sqrt(l))
```
